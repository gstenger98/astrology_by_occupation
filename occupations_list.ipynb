```python
from pymongo import MongoClient
import pprint

import copy
import pandas as pd

import requests

from bs4 import BeautifulSoup
```

```python
occupations_url = 'https://famouspeople.astro-seek.com/filter_occupation/'
occupations_r = requests.get(occupations_url)
```

```python
client = MongoClient('localhost', 27017)
db = client.metroid
pages = db.pages

pages.insert_one({'html': occupations_r.content})
```

```python
print(db)
```

```python
soup = BeautifulSoup(occupations_r.content, 'html')
print(soup.prettify())
```

```python
#div = soup.find('div', {'style': 'width: 170px; float: left; overflow: hidden;'})
#table = div.find('name')
#print(div)

div = soup.find('div', {'class': 'tab_content', 'id': 'tab8', 'style': 'display: block;'})

#table = div.find('name')

name = div.find('optgroup')
#<select name="povolani" style="">

#<div class="tab_content" id="tab8" style="display: block;">



#print(div.prettify())
print(name)
```

```python
print(name.prettify())
```

```python
'''
<optgroup label="Occupation">
 <option value="astrologers">
  Astrologers
 </option>
 <option value="chemists">
  Chemists
 </option>
'''

rows = name.find_all('option')
all_rows = []

for row in rows[0:]:
    #new_row = row.find_all('option values').text.strip()
    new_row = row.text.strip()
    all_rows.append(new_row)

pprint.pprint(all_rows)
```

```python
occupations_list = [_.lower() for _ in all_rows]
type(occupations_list)
#print(occupations_list)
```

At this point we have our occupations_list. Now let's turn the list into urls.

```python
url_beginning = "https://famouspeople.astro-seek.com/filter_occupation/?narozeni_stat=&povolani="
url_ending = "&narozeni_stat=&pohlavi=&razeni="
    
occupations_urls = [url_beginning + occupation_title + url_ending for occupation_title in occupations_list]
print(occupations_urls)
#for _ in occupations_urls:
    #print(_)
```

We have our urls! Now what we do is scrape the data from each of these urls.

```python
df = pd.DataFrame(occupations_urls)
df.head()
```

```python
df.to_csv('occupations_urls.csv')
```

```python
df_new = pd.DataFrame(occupations_urls, columns=['urls'])
df_new.head()
```

```python
df_new['names'] = occupations_list
df_new.head()
```

```python
#df_new.to_csv('occupations_urls_and_names.csv', index=False)
df_new.reset_index(drop=True, inplace=True)
df_new.head()
```

```python
df_new.to_csv('occupations_urls_and_names.csv', index=False)
```
