```python
import warnings
warnings.filterwarnings('ignore')

from pymongo import MongoClient
import pprint

import copy
import pandas as pd

import requests

from bs4 import BeautifulSoup
```

```python
occupations_url = 'https://famouspeople.astro-seek.com/filter_occupation/'
occupations_r = requests.get(occupations_url)

client = MongoClient('localhost', 27017)
db = client.metroid
pages = db.pages

pages.insert_one({'html': occupations_r.content})

soup = BeautifulSoup(occupations_r.content, 'html')
```

```python
div = soup.find('div', {'class': 'tab_content', 'id': 'tab8', 'style': 'display: block;'})
name = div.find('optgroup', {'label': 'Occupation'})
#name = div.find('optgroup', {'label': 'Sports'})
#name = div.find('optgroup', {'label': 'Arts'})
#name = div.find('optgroup', {'label': 'Other'})


rows = name.find_all('option')
all_rows = []

for row in rows[0:]:
    #new_row = row.find_all('option values').text.strip()
    new_row = row.text.strip()
    all_rows.append(new_row)




occupations_list = [_.lower() for _ in all_rows]

occupations_list_mod = []
for elem in occupations_list:
    new_elem = elem.replace(', ', '-')
    newest_elem = new_elem.replace(' ', '-')
    occupations_list_mod.append(newest_elem)


url_beginning = "https://famouspeople.astro-seek.com/filter_occupation/?narozeni_stat=&povolani="
url_ending = "&narozeni_stat=&pohlavi=&razeni="

occupations_urls_mod = [url_beginning + occupation_title + url_ending for occupation_title in occupations_list_mod]

df_occupation = pd.DataFrame(occupations_urls_mod, columns=['urls'])
df_occupation['names'] = occupations_list_mod

#df_sports = pd.DataFrame(occupations_urls_mod, columns=['urls'])
#df_sports['names'] = occupations_list_mod

#df_arts = pd.DataFrame(occupations_urls_mod, columns=['urls'])
#df_arts['names'] = occupations_list_mod

#df_other = pd.DataFrame(occupations_urls_mod, columns=['urls'])
#df_other['names'] = occupations_list_mod
```
